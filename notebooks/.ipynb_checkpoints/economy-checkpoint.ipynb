## Data & cleaning/processing

### Cities
We'll use the same list of cities that we did in the climate section, which is from [Wikipedia](https://en.wikipedia.org/wiki/List_of_metropolitan_statistical_areas). 

```python
cities = pd.read_csv(
    '../data/raw/metros.csv',
    header=0,
    usecols=[1, 2, 3, 4],
    names=['Metro', '2017 pop', '2010 pop', 'pop change']
)
cities.head()
```

Before we can do anything interesting with these data, we're going to have to change cast them to a workable format (right now they're stored as strings, we need them as numbers).

```python
cities['2017 pop'] = cities.apply(
    lambda city: int(city['2017 pop'].replace(',', '')),
    axis=1
)
cities['2010 pop'] = cities.apply(
    lambda city: int(city['2010 pop'].replace(',', '')),
    axis=1
)
cities['pop change'] = cities.apply(
    lambda city: float(
        city['pop change'].replace(u'\u2212','-').replace('+', '').replace('%', '')
        # the data uses the unicode minus sign instead of regular dashes on negatives
    ) / 100,
    axis=1
)
cities.dtypes
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,6), dpi=150)
cities.hist(column='2017 pop', ax=ax)
plt.show()
```

What cities are growing the fastest?

```python
cities.sort_values('pop change', ascending=False).head(30)
```

```python
cities.hist(column='pop change')
plt.show()
```

It's hard to say whether it's better to live in a big city or a small one, but what we can do is combine the population data with GDP. [GDP data are from the Bureau of Economic Analysis](https://www.bea.gov/data/gdp/gdp-metropolitan-area). Note that the values for GDP are in millions of dollars.

```python
gdp = pd.read_csv(
    '../data/raw/economy/nominal-gdp.csv',
    header=2,
    usecols=[0,6,7],
    names=['Metro', '2017 GDP', '2017 Rank']
)
# the first row is an aggregate of all rows,
# the 2 rows are not data (contain information about the source)
gdp = gdp.iloc[1:383,:]
gdp['2017 GDP'] = gdp.apply(lambda x: int(str(x['2017 GDP']).replace(',', '')), axis=1)
gdp['2017 Rank'] = gdp.apply(lambda x: int(str(x['2017 Rank'])), axis=1)
gdp = gdp.sort_values('2017 Rank')
gdp.head()
```

Let's join the GDP data with the cities table. Before we do that, we'll have to remove the 'MSA' that's at the end of every metro name in the cities table.

```python
cities['Metro'] = cities.apply(
    lambda x: x['Metro'][0 : x['Metro'].find(' MSA')],
    axis=1
)
cities.head()
```

```python
cities = cities.merge(gdp, on='Metro', how='inner')
cities.head()
```

Now we can create a new column for GDP per capita.

```python
cities['2017 GDP per capita'] = cities.apply(
    lambda x: (x['2017 GDP'] * (10 ** 6)) / x['2017 pop'],
    axis = 1
)
cities.head()
```

```python
cities.hist(column='2017 GDP per capita')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Looks like there are a few outliers where the GDP per capita is enormous. Let's take a look at some of the richest cities.

```python
cities.sort_values('2017 GDP per capita', ascending=False).head(30)
```

All of these cities could be considered to have an outstanding economy by this simple measure. To take it a step further, let's calculate real GDP per capita growth from 2010 to 2017.

But first, I'm interested whether GDP per capita increases with the size of the city. Does a bigger city unlock economic possibilities unavailable to a smaller city?

```python
sb.scatterplot(
    x=cities['2017 pop'],
    y=cities['2017 GDP per capita']
)
plt.show()
```
